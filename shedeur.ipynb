```python
import matplotlib.pyplot as plt
import pandas as pd

# Data from previous turn
data = {
    "Team": [
        "LA Rams", "New England Patriots", "Seattle Seahawks", "Detroit Lions", "Buffalo Bills",
        "Jacksonville Jaguars", "Dallas Cowboys", "Indianapolis Colts", "Chicago Bears", "San Francisco 49ers",
        "Baltimore Ravens", "Cincinnati Bengals", "Houston Texans", "Denver Broncos", "Pittsburgh Steelers",
        "Green Bay Packers", "NY Giants", "Tampa Bay Buccaneers", "Philadelphia Eagles", "LA Chargers",
        "Kansas City Chiefs", "Washington Commanders", "Arizona Cardinals", "Atlanta Falcons", "Miami Dolphins",
        "Minnesota Vikings", "Carolina Panthers", "New Orleans Saints", "NY Jets", "Tennessee Titans",
        "Cleveland Browns", "Las Vegas Raiders"
    ],
    "PPG": [
        30.5, 28.8, 28.4, 28.3, 28.3, 27.9, 27.7, 27.4, 25.9, 25.7,
        24.9, 24.4, 23.8, 23.6, 23.4, 23.0, 22.4, 22.4, 22.3, 21.6,
        21.3, 20.9, 20.9, 20.8, 20.4, 20.2, 18.3, 18.0, 17.6, 16.7,
        16.4, 14.2
    ]
}

# Official Team Primary Colors (Hex codes)
team_colors = {
    "LA Rams": "#003594",
    "New England Patriots": "#002244",
    "Seattle Seahawks": "#002244",
    "Detroit Lions": "#0076B6",
    "Buffalo Bills": "#00338D",
    "Jacksonville Jaguars": "#006778",
    "Dallas Cowboys": "#003594",
    "Indianapolis Colts": "#002C5F",
    "Chicago Bears": "#0B162A",
    "San Francisco 49ers": "#AA0000",
    "Baltimore Ravens": "#241773",
    "Cincinnati Bengals": "#FB4F14",
    "Houston Texans": "#03202F",
    "Denver Broncos": "#002244",
    "Pittsburgh Steelers": "#000000",
    "Green Bay Packers": "#203731",
    "NY Giants": "#0B2265",
    "Tampa Bay Buccaneers": "#D50A0A",
    "Philadelphia Eagles": "#004C54",
    "LA Chargers": "#0080C6",
    "Kansas City Chiefs": "#E31837",
    "Washington Commanders": "#5A1414",
    "Arizona Cardinals": "#97233F",
    "Atlanta Falcons": "#A71930",
    "Miami Dolphins": "#008E97",
    "Minnesota Vikings": "#4F2683",
    "Carolina Panthers": "#0085CA",
    "New Orleans Saints": "#D3BC8D",
    "NY Jets": "#125740",
    "Tennessee Titans": "#0C2340",
    "Cleveland Browns": "#311D00",
    "Las Vegas Raiders": "#A5ACAF" # Using Silver to distinguish from Steelers
}

df = pd.DataFrame(data)
df["Color"] = df["Team"].map(team_colors)

# Sort by PPG for the bar graph
df_sorted = df.sort_values(by="PPG", ascending=True)

# Create the bar graph
plt.figure(figsize=(14, 12))
bars = plt.barh(df_sorted["Team"], df_sorted["PPG"], color=df_sorted["Color"], edgecolor='black', linewidth=0.5)

plt.xlabel("Points Per Game (PPG)", fontsize=12)
plt.ylabel("NFL Team", fontsize=12)
plt.title("NFL Team Points Per Game - 2025 Regular Season", fontsize=16, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Add value labels
for index, value in enumerate(df_sorted["PPG"]):
    plt.text(value + 0.3, index, f"{value:.1f}", va='center', fontweight='bold')

plt.xlim(0, 35) # Give space for labels
plt.tight_layout()
plt.savefig("nfl_2025_ppg_colored.png")
print("Colored graph saved successfully.")
```
